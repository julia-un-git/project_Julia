```python
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
```

```python
nltk.download('vader_lexicon')
```

```python
nltk.download(["stopwords"])
```

```python
from nltk.tokenize import word_tokenize
```

```python
stopwords = nltk.corpus.stopwords.words("english")
```

```python
df = pd.read_csv('clear_data.csv')
```

```python
df = df[['product_id','review_title','review_content', 'rating']]
df
```

```python
df.dtypes
```

```python
df["review_title"] = df["review_title"].astype("string")
df["review_content"] = df["review_content"].astype("string")
```

```python
df["review_title"] = df["review_title"].apply(lambda x: x.lower())
df["review_content"] = df["review_content"].apply(lambda x: x.lower())
```

```python
df['review_title'] = df['review_title'].str.replace(",",' ')
df['review_content'] = df['review_content'].str.replace(",",' ')
df['review_title'] = df['review_title'].str.replace(".",' ')
df['review_content'] = df['review_content'].str.replace(".",' ')
df['review_title'] = df['review_title'].str.replace("/",' ')
df['review_content'] = df['review_content'].str.replace("/",' ')
```

```python
def remove_words(review):
    # Токенизируем отзыв на слова
    tokens = word_tokenize(review)
    # Удаляем слова из списка
    tokens = [word for word in tokens if word not in stopwords]
    # Объединяем токены обратно в отзыв
    new_review = ' '.join(tokens)
    return new_review
```

```python
df['review_title'] = df['review_title'].apply(remove_words)
df['review_content'] = df['review_content'].apply(remove_words)
```

```python
df
```

```python
sia = SentimentIntensityAnalyzer()
```

```python
df['sentiment'] = df['review_title'].apply(lambda x: sia.polarity_scores(x)['compound'])
df['sentiment2'] = df['review_content'].apply(lambda x: sia.polarity_scores(x)['compound'])
```

```python
df
```

Оценка настроений представлена в виде компаунд ('compound')-показателя, который может принимать значения от -1 (отрицательное настроение) до 1 (положительное настроение). Значения близкие к 0 считаются нейтральными.
