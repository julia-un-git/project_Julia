Для работы с данными приведем их к нужному типу. 
Необходимо удалить значок валюты перед ценами и убрать разделитель групп разрядов ",".
Далее колонки с ценами и рейтингами приводим к типу float.

```python
amazon['discounted_price'] = amazon['discounted_price'].str.replace("₹",'')
amazon['discounted_price'] = amazon['discounted_price'].str.replace(",",'')
```

```python
amazon.head (10)
```

```python
amazon['discounted_price'] = amazon['discounted_price'].astype('float64')
```

```python
amazon.head (5)
```

```python
amazon['actual_price'] = amazon['actual_price'].str.replace("₹",'')
amazon['actual_price'] = amazon['actual_price'].str.replace(",",'')
amazon['actual_price'] = amazon['actual_price'].astype('float64')
```

```python
amazon.head (5)
```

```python
amazon['discount_percentage'] = amazon['discount_percentage'].str.replace('%','').astype('float64')
amazon['discount_percentage']
```

```python
# amazon['rating'] =amazon['discount_percentage'].str.replace('%','')
```

Разделяем категорию на подтипы, для удобства работы с ними

```python
df_category = df['category'].str.split('|', expand=True).rename(columns={0:'category_1', 1:'category_2', 2:'category_3'})
df_category.head()
```

```python
#уникальные значения первой категории
df_category['category_1'].value_counts()
```

```python
df_category.isnull().sum() # обращаем внимание, что 3 категория присутствует не у всех продуктов
```

```python
df["category_main"] = df_category["category_1"]
df["category_sub"] = df_category["category_2"]
```

```python
df
```

```python
df['type_product'] = df['category'].astype(str).str.split('|').str[-1] #сохраним последнюю категорию - определяющую тип изучаемого продукта
```

```python
df.drop('category', axis=1, inplace=True) #удаляем исходный столбец категорий
```
